```python
import ast
import numpy as np
```

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
```

```python
f = open("win_256 DQN.txt", 'r')
data = f.read()
data = ast.literal_eval(data)
print(type(data))
print(sum(data))
print(len(data))
f.close()
```

```python
Y1 = moving_average(data, 2000)
X1 = range(len(Y1))
```
