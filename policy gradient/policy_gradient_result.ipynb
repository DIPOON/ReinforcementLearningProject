```python
# Initialization goes here
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os
import pandas as pd
import math
from matplotlib import rc
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

def smooth(y,factor):
    if type(y)!=list:
        y = list(y)
    return pd.Series(y).rolling(window=factor).mean()#[factor:]
```

```python
import numpy as np
import pandas as pd
import IPython
import matplotlib.pyplot as plt
import pickle 
epochs = []
steps = []
train_r = []
loss = []
winrate = []

stats = pickle.load(open("condensed_6x6_CNN/stats.p", "rb"))
for stat in stats:
    steps.append(stat[1])
    train_r.append(stat[2]/1.5)
    loss.append(stat[3])
    winrate.append(stat[4])

smooth_loss = smooth(loss, 200)
smooth_winrate = smooth(winrate, 200)
```
